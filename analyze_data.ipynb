```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the Netflix dataset
netflix_df = pd.read_csv("netflix_titles.csv")

# Filter the dataset to only include movies
netflix_movies = netflix_df[netflix_df['type'] == 'Movie']

# Select a subset of columns to use in the scatter plot
netflix_movies_col_subset = netflix_movies[['release_year', 'duration']]

# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'])

# Create a title
plt.title('Movie Duration by Year of Release')

# Show the plot
plt.show()
```

```python
import pandas as pd

# Create a DataFrame from the dictionary
movie_dict = {
    'years': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'durations': [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
}

durations_df = pd.DataFrame(movie_dict)

# Convert 'durations' column to numeric values
durations_df['durations'] = pd.to_numeric(durations_df['durations'], errors='coerce')

# Filter for durations shorter than 60 minutes
short_movies = durations_df[durations_df['durations'] < 100]

# Print the first 20 rows of short_movies
print(short_movies.head(20))
```

```python
import pandas as pd

# Define an empty list
colors = []

# Check if durations_df is a DataFrame
if isinstance(durations_df, pd.DataFrame):
    # Check if 'genre' column exists
    if 'genre' in durations_df.columns:
        # Iterate over rows of durations_df
        for idx, row in durations_df.iterrows():
            if row['genre'] == 'Children':
                colors.append('red')
            elif row['genre'] == 'Documentaries':
                colors.append('blue')
            elif row['genre'] == 'Stand-Up':
                colors.append('green')
            else:
                colors.append('black')
    else:
        print('Column does not exist.')
else:
    print('durations_df is not a DataFrame.')
```
